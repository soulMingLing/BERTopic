```python
import re
# 定义一个函数，用于读取文件并分句
def read_and_split_sentences(file_path, output_file_path):
    # 打开文件并读取内容
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # 使用正则表达式分句，句子以句号、问号或感叹号结束
    sentences = re.split(r'(?<=[。！？])\s*', content)
    
    # 移除空字符串
    sentences = [sentence for sentence in sentences if sentence]
    
    # 将分句后的内容写入新文件
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        i=0
        for sentence in sentences:
            i=i+1
            if i == len(sentences):  # 最后一行不加换行
                output_file.write(sentence)
            else:
                output_file.write(sentence + "\n")


# 使用函数
input_file_path = '党史数据库.txt'  # 输入文件路径
output_file_path = '党史数据库-3.txt'  # 输出文件路径
read_and_split_sentences(input_file_path, output_file_path)
```
